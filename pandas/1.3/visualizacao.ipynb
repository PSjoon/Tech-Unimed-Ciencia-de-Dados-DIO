```python
import pandas as pd 
```

```python
df = pd.read_excel("./Aracaju.xlsx")
```

```python
#contagem de repeticoes 
df['LojaID'].value_counts(ascending=False)
```

```python
#grafico barras pela biblioteca matplotlib 
df['LojaID'].value_counts(ascending=True).plot.bar()
```

```python
#grafico barras horizontais
df['LojaID'].value_counts(ascending=True).plot.barh()
```

```python
#grafico pizza
df['LojaID'].value_counts(ascending=True).plot.pie()
```

```python
#personalizacao do gradico com alteracao de title e color
#para personalizacoes mais precisas, consultar documentacao
import matplotlib.pyplot as plt
from matplotlib.pyplot import title
df["Cidade"].value_counts().plot.bar(title="Total Vendas Cidade", color="red")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas")
```
