```python
import pandas as pd 
```

```python
df = pd.read_excel("./Aracaju.xlsx")

#conversao de valores a data
df['Data'] = pd.to_datetime(df["Data"])
```

```python
df.groupby(df["Data"].dt.year)["Receitas"].sum()
df['Ano_venda'] = df['Data'].dt.year
```

```python
#retorno de diferenca entre datas
df['doferenca'] = df["Data"] - df['Data'].min()
```

```python
#retorno de trimestre
df['trimestre'] = df["Data"].dt.quarter
```

```python
#filtragem de dados
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 1)]
```
